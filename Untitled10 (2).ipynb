```python
G = DihedralGroup(20)

# Виведемо всі нормальні підгрупи групи G
normal_subgroups = G.normal_subgroups()
print("Нормальні підгрупи групи D20:")
for N in normal_subgroups:
    print(N)

# Функція для побудови ендоморфізму фі вручну
def create_endomorphism(G, N):
    """
    Створює ендоморфізм для групи G, ядро якого дорівнює підгрупі N.
   
    """
    # Визначаємо ендоморфізм phi: кожен елемент g відображається так, щоб
    # якщо g в підгрупі N, то phi(g) = нейтральний елемент, інакше g залишається без змін.
    
    def phi(g):
        if g in N:
            return G.identity()  # нейтральний елемент групи G
        else:
            return g  

    # Ручна перевірка ядра ендоморфізму, знайдемо всі елементи, які відображаються в нейтральний елемент
    kernel_elements = [g for g in G if phi(g) == G.identity()]
    
    # Створюємо підгрупу з елементів ядра
    kernel = G.subgroup(kernel_elements)
    
    return kernel, phi

# Перевірка для кожної нормальної підгрупи
for N in normal_subgroups:
    print(f"Перевірка для підгрупи {N}:")
    
    # Створення ендоморфізму для поточної нормальної підгрупи N
    kernel, phi_map = create_endomorphism(G, N)
    
    # Виводимо ядро
    print(f"Ядро ендоморфізму: {kernel}")
    
    # Перевіряємо, чи ядро дорівнює підгрупі N
    if kernel == N:
        print(f"Ядро ендоморфізму дорівнює підгрупі N.")
    else:
        print(f"Ядро ендоморфізму НЕ дорівнює підгрупі N.")
```

```python
def analyze_group(G, name):
   
    print(f"Група: {name}")
    print(f"Кількість елементів у групі |G| = {G.order()}")
    
    # Центр групи
    center = G.center()
    print(f"Центр групи Z(G): {center}")
    print(f"Кількість елементів у центрі: {center.order()}")
    
    # Класи спряженості
    conjugacy_classes = G.conjugacy_classes()
    print(f"Кількість класів спряженості: {len(conjugacy_classes)}")
    print("Представники класів спряженості:")
    for cls in conjugacy_classes:
        print(f"  {cls.representative()} (розмір класу: {len(cls)})")
    
    # Формула для кожного елемента групи на основі його центрального елементу
    formula = " + ".join([f"|G| / |C_G({g})| = {G.order()} / {G.centralizer(g).order()}" 
                          for g in G.conjugacy_classes_representatives()])
    print(f"Формула для групи {name}: {formula}")
    
    # Виведення готової формули для кожної групи
    if name == "C11":
        print(f"Формула для C11: |C11| = 11, кількість класів спряженості = 11")
    elif name == "D13":
        print(f"Формула для D13: |D13| = 26, кількість класів спряженості = 13")
    elif name == "Q8":
        print(f"Формула для Q8: |Q8| = 8, кількість класів спряженості = 5")
    
    # Перевірка формули класів
    sizes = [G.order() / G.centralizer(g).order() for g in G.conjugacy_classes_representatives()]
    print(f"Перевірка формули класів: {G.order()} == {sum(sizes)} -> {G.order() == sum(sizes)}")
    print("-" * 50)

# Аналіз груп

# 1. Циклічна група C11
C11 = CyclicPermutationGroup(11)
analyze_group(C11, "C11")

# 2. Діедральна група D13
D13 = DihedralGroup(13)
analyze_group(D13, "D13")

# 3. Кватерніонна група Q8
Q8 = QuaternionGroup()
analyze_group(Q8, "Q8")
```

```python
# Створення графу гіперкуба C_4
G = graphs.CubeGraph(4)


print("Граф G:")
G.show()


automorphisms = G.automorphism_group()
print("Група автоморфізмів графу:", automorphisms)

# Дія групи автоморфізмів на граф
print("Розмір групи автоморфізмів:", automorphisms.order())

# Орбіти групової дії на вершинах
orbits = automorphisms.orbits()
print("Орбіти групової дії на вершинах:")
for orbit in orbits:
    print(orbit)

# Знайдемо стабілізатор довільної вершини (наприклад, вершина з індексом 0)
vertex = G.vertices()[0]  # Отримуємо першу вершину (індекс 0)
stabilizer = automorphisms.stabilizer(vertex)
print(f"\nСтабілізатор вершини {vertex}:", stabilizer)

# Орбіти стабілізатора
stabilizer_orbits = stabilizer.orbits()
print("Орбіти стабілізатора:")
for orbit in stabilizer_orbits:
    print(orbit)

# Дія стабілізатора на весь граф
print("Розмір стабілізатора:", stabilizer.order())
```

```python
# Функція для перевірки групи автоморфізмів графа
def verify_group_isomorphism(graph, expected_group_order, group_label):
    automorphism_group = graph.automorphism_group()
    print(f"Група автоморфізмів графа {group_label}: {automorphism_group.structure_description()}")
    
    # Перевірка на ізоморфізм
    if len(automorphism_group) == expected_group_order:
        print(f"Група автоморфізмів графа {group_label} має такий самий порядок, як у групи порядку {expected_group_order}: True")
    else:
        print(f"Група автоморфізмів графа {group_label} має такий самий порядок, як у групи порядку {expected_group_order}: False")


# 1) Для A2 (граф з 2 вершинами, циклічний граф)
vertices_A2 = 2
edges_A2 = [(0, 1)]  # Граф C2
graph_A2 = Graph()
graph_A2.add_edges(edges_A2)

# Відображення графа для A2
graph_A2.show()

# Перевірка групи автоморфізмів для A2 (C2)
verify_group_isomorphism(graph_A2, 2, "A2")  # Очікуємо групу C2 порядку 2

# 2) Для A3 (граф з 3 вершинами, тріангулярний граф)
vertices_A3 = 3
edges_A3 = [(0, 1), (1, 2), (2, 0)]  # Трикутник K3
graph_A3 = Graph()
graph_A3.add_edges(edges_A3)

# Відображення графа для A3
graph_A3.show()

# Перевірка групи автоморфізмів для A3 (S3)
verify_group_isomorphism(graph_A3, 6, "A3")  # Очікуємо групу S3 порядку 6

# 3) Для A4 (граф з 4 вершинами)
vertices_A4 = 4
edges_A4 = [(0, 1), (1, 2), (2, 3), (3, 0), (0, 2), (1, 3)]  # Граф для A4 (тетраедр)
graph_A4 = Graph()
graph_A4.add_edges(edges_A4)

# Відображення графа для A4
graph_A4.show()

# Перевірка групи автоморфізмів для A4 (S4)
verify_group_isomorphism(graph_A4, 24, "A4")  # Очікуємо групу S4 порядку 24

# 4) Для A5 (граф з 5 вершинами, пентагон або повний граф K5)
vertices_A5 = 5
edges_A5 = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 0)]  # (C5)
graph_A5 = Graph()
graph_A5.add_edges(edges_A5)

# Відображення графа для A5
graph_A5.show()

# Перевірка групи автоморфізмів для A5 (D5)
verify_group_isomorphism(graph_A5, 10, "A5")  # Очікуємо групу D5 порядку 10
```
